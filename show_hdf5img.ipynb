```python
import h5py
import ipyplot
from PIL import Image
```

```python
data = None
filename = "datasets/shapenet/data_chair.hdf5"
f = h5py.File(filename, "r")
```

```python
# List all groups
keys = list(f.keys())
print("Example keys (sampling first 10-keys): %s\n" % keys[:10])
```

```python
group = f['11c9c57efad0b5ec297936c81e7f6629_4_10']
```

```python
list(group.keys())
```

```python
group['pose'][()]
```

```python
data = group['image'][()]
```

```python
data
```

```python
Image.fromarray(data)
```

```python
num_img = 200
max_img = num_img
count_img = 0
img_arr = []

for key in keys:
    if count_img == num_img:
        break
        
    each_group = f[key]
    data = each_group['image'][()]
    img_arr.append(data)
    count_img = count_img + 1

ipyplot.plot_images(img_arr, max_images=max_img, img_width=150)
```
